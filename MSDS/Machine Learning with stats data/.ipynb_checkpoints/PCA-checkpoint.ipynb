# Principal component analysis: Analyzing voting behavior in the US Congress

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the datasets
votes_df = pd.read_csv('rollcall-votes.csv')
members_df = pd.read_csv('rollcall-members.csv')

#Standardizing the vote data since PCA is affected by scale
scaler = StandardScaler()
votes_scaled = scaler.fit_transform(votes_df)
```

# TASK 1: Running PCA, Visualization, Decide the number of principal components

```python
# First, let's run PCA on the standardized voting data without specifying the number of components
# to retain all of them, which allows us to see the full variance explained by each component.

pca_all = PCA()
principal_components_all = pca_all.fit_transform(votes_scaled)

# Creating the scree plot to show the explained variances of all principal components
# scree plot will be used to visualize the importance of each principal component
plt.figure(figsize=(15, 5))
plt.plot(pca_all.explained_variance_ratio_, marker='o', linestyle='--')
plt.title('Scree Plot for PCA of Voting Data')
plt.xlabel('Number of Components')
plt.ylabel('Variance Explained')
plt.grid(True)
plt.show()

# Creating a cumulative variance plot to help decide on the number of PCs to use
plt.figure(figsize=(15, 5))
plt.plot(pca_all.explained_variance_ratio_.cumsum(), marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Variance Explained')
plt.axhline(y=0.85, color='r', linestyle='-')  # 85% threshold line for reference
plt.grid(True)
plt.show()

# The first plot is the scree plot which shows how much variance is explained by each component.
# The second plot shows the cumulative variance explained as we add more components.
# These plots help understand the trade-off between the number of components and the amount of variance explained.

# Now let's find out the number of components that explain 85% of the variance
cumulative_variance = pca_all.explained_variance_ratio_.cumsum()
n_components_for_85 = (cumulative_variance < 0.85).sum() + 1  # Add 1 due to zero-based indexing

n_components_for_85
```

## <u>Scree Plot</u>

### It is evident from the plot that the first few components explain a significant amount of variance, with a sharp drop-off as we move to higher-numbered components. 
### This plot helps identify where the 'elbow' is, which is the point after which additional components contribute less to the explanation of variance. 
### The 'elbow' method is a heuristic used in determining the number of components to keep.

## <u>Cumulative Variance Plot</u>

### The cumulative variance plot is a visualization of the cumulative sum of explained variances by the components. 
### The horizontal red line marks the 85% threshold, which is a common cut-off for sufficient variance explanation in many applications. 
### The plot indicates that 122 components are needed to explain approximately 85% of the variance in the data.


## <u>Interpretation</u>

### With 122 components explaining 85% of the variance, it suggests that the voting patterns are multifaceted.
### While the first few components capture the most significant patterns of variance, additional components are needed to capture the subtler, yet substantial, aspects of the data.

### The analysis shows that while it's possible to reduce the dimensionality of the data significantly, a relatively large number of dimensions are still required to capture the majority of the information in the original data. 
### This highlights the complexity of the voting behavior in the dataset and suggests that many nuances would be lost if the dimensionality were reduced too drastically.

```python
# Run PCA again with 122 components to capture 85% of variance
pca_122 = PCA(n_components=n_components_for_85)
principal_components_122 = pca_122.fit_transform(votes_scaled)

#DataFrame for the first two principal components for visualization
pca_2_df = pd.DataFrame(principal_components_122[:, :2], columns=['PC1', 'PC2'])

# Merge the PCA results with the members info
pca_members_df = pd.concat([pca_2_df, members_df], axis=1)

# Plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='party', data=pca_members_df, palette='viridis')
plt.title('PCA of Roll Call Votes by Party')
plt.xlabel('Principal Component 1 (PC1) - Represents the most variance')
plt.ylabel('Principal Component 2 (PC2) - Represents the second most variance')
plt.legend(title='Party', loc='best')
plt.grid(True)
plt.show()
```

## <u>PC1 vs PC2</u>

### Separation by Party Affiliation: The plot displays a discernible pattern of separation along PC1, which is the horizontal axis representing the most variance. We can see that:
### The 'R' (presumably Republican) party members tend to have positive scores on PC1.
### The 'D' (presumably Democrat) party members tend to have negative scores on PC1.
### Members labeled as 'DR' are situated in the middle to the right side of the plot, which could represent independent members or those with mixed party affiliations.

### Polarization: The clear separation along PC1 may suggest a strong polarization in voting behavior between the two major parties, with each party clustering on opposite sides of the plot. This is consistent with the expected political divisions in a two-party system.

### Outliers: There are a few data points with extreme values on PC2, both positive and negative. These could be legislators who have voting records that are highly atypical compared to their colleagues, perhaps voting frequently against party lines or on specific issues that are not as strongly correlated with the main axis of party division.
### Potential Subgroup Identification: If we look at the distribution of 'DR' points, they do not form a separate cluster but rather are placed within the spread of the other parties, suggesting that these members' voting behavior is somewhat aligned with the major parties but with some variance.
### Dimensionality of Voting Behavior: While PC1 seems to capture a significant variance along party lines, PC2 might be capturing a more issue-specific aspect of voting that doesn't separate along the same axis as party affiliation.


```python
# factor loadings for the first two principal components
loadings = pd.DataFrame(pca_122.components_[:2], columns=votes_df.columns, index=['PC1', 'PC2']).T
loadings['abs_PC1'] = loadings['PC1'].abs()
loadings['abs_PC2'] = loadings['PC2'].abs()

# top 10 loadings for PC1 and PC2
top_loadings_PC1 = loadings.sort_values(by='abs_PC1', ascending=False).head(10)
top_loadings_PC2 = loadings.sort_values(by='abs_PC2', ascending=False).head(10)

# results
print("Top 10 loadings for PC1:")
display(top_loadings_PC1)
print("\nTop 10 loadings for PC2:")
display(top_loadings_PC2)
```

# TASK 2: Interpret PC score for each legislator

```python
# add the principal component scores to the members dataframe
members_df['PC1_score'] = principal_components_all[:, 0]  # PC1 scores

# We sort the dataframe based on the PC1 score to find the legislators with the largest and smallest scores
sorted_members = members_df.sort_values(by='PC1_score', ascending=False)

#legislators with the largest PC1 scores (top 5)
top_5_PC1_scores = sorted_members.head(5)

#legislators with the smallest PC1 scores (bottom 5)
bottom_5_PC1_scores = sorted_members.tail(5)

display(top_5_PC1_scores[['member', 'state', 'party', 'PC1_score']])
display(bottom_5_PC1_scores[['member', 'state', 'party', 'PC1_score']])
```

## <u>Top Legislators by PC1 Score</u>

### The legislators with the highest PC1 scores are predominantly from the Republican party (R), as indicated by the data. 
### This suggests that the first principal component may be capturing a dimension of variance that aligns with a particular aspect of conservative voting behavior or perhaps specific issues that these legislators tend to vote similarly on. 
### The fact that these legislators stand out on PC1 could mean they are more consistent or extreme in this behavior compared to their peers.

## <u>Bottom Legislators by PC1 Score</u>

### Conversely, the legislators with the lowest PC1 scores are from the Democratic party (D), indicating that PC1 also captures variance in voting behavior that aligns with liberal voting patterns. 
### Legislators with the most negative scores on PC1 may represent a liberal equivalent to the conservative pattern captured by the positive end of PC1.

## <u>Interpretation of PC1</u>

### The first principal component appears to be a strong indicator of partisan voting behavior, distinguishing between Republican and Democratic patterns. 
### This is not entirely surprising given the polarized nature of US politics

## <u>Surprises and Learning Points</u>

### The result of this analysis may shock some people, especially regarding the magnitude of scores and distinctiveness between the two parties.
### This means that one single principle component can account for a good proportion of partisan differences in electoral behaviour.


## <u>Legislators with Extreme PC1 Scores</u>

### The legislators with the largest and smallest PC1 scores could be of particular interest for a more detailed analysis. 
### Their voting records could be examined to understand which specific issues they are more likely to vote for or against, and whether their scores are a result of a few key votes or a general trend across many votes.

# TASK 3: Vote Loadings for the First PC

```python
#loadings for the first principal component
loadings_PC1 = pca_all.components_[0]

#plot for the loadings
plt.figure(figsize=(15, 5))
plt.bar(range(len(loadings_PC1)), loadings_PC1, align='center')
plt.title('Loadings for the First Principal Component (PC1)')
plt.xlabel('Votes')
plt.ylabel('Loadings')
plt.show()
```

## <u>How to Interpret the Loadings</u>:

### Positive Loadings: Votes with positive loadings on PC1 suggest that legislators with a high score on this component tend to vote 'yea' on these issues. 
### The higher the loading, the more strongly a 'yea' vote on that issue is associated with the pattern captured by PC1.
### Negative Loadings: Votes with negative loadings suggest that legislators with a high score on PC1 tend to vote 'nay' on these issues. 
### A lower (more negative) loading implies a stronger association with a 'nay' vote in the pattern captured by PC1.
### Magnitude of Loadings: The magnitude (absolute value) of the loading indicates the strength of the association between the vote and PC1. 
### Larger magnitudes (both positive and negative) mean that the vote has a stronger relationship with the variance captured by PC1.

## <u>Observations from the Plot</u>:

### Variation in Loadings: There is considerable variation in the loadings, with both positive and negative values, suggesting that PC1 captures a complex pattern of voting behavior that includes both 'yea' and 'nay' votes.
### Significant Votes: Votes that have higher positive or negative loadings are the ones that contribute most to PC1. 
### These votes could be considered significant in differentiating between the voting behaviors of legislators and could be further analyzed to understand the specific issues that are driving the pattern of variance captured by PC1.

# TASK 4: Analyse vote loadings for the Second PC

```python
#loadings for the second principal component
pca = PCA()
principal_components = pca.fit_transform(votes_scaled)
loadings_PC2 = pca.components_[1]

# Plotting the loadings
plt.figure(figsize=(20, 4))
plt.bar(range(len(loadings_PC2)), loadings_PC2, align='center')
plt.xlabel('Votes')
plt.ylabel('Loading on PC2')
plt.title('Vote Loadings on the Second Principal Component (PC2)')
plt.show()
```

## <u>Characterization of Votes with Large Loadings on PC2</u>:

### Votes with Large Positive Loadings: These votes are associated with a specific pattern that is different from the one captured by PC1. 
### This pattern could be related to issues that differentiate legislators in a way that is not strictly partisan but could be related to other factors such as ideology, geography, or specific interests.

### Votes with Large Negative Loadings: These votes represent another pattern that is also distinct from PC1. 
### Legislators who score low on PC2 might consistently vote against these issues, indicating a separate dimension of disagreement or agreement that is independent of the primary partisan divide.


## <u>Insights</u>:

### Votes with large loadings on PC2 can be characterized as those that capture variance in voting behavior not explained by PC1. 
### While PC1 might primarily capture the partisan divide, PC2 could be revealing other underlying factors influencing voting behavior. 
### This component might represent factors such as regional issues, the interests of specific constituencies, or coalitions on certain legislation.

# TASK 5: Analyze the scores associated with the second principal component for each legislator

```python
# scores for the second principal component into the members dataframe
members_df['PC2_score'] = principal_components[:, 1]  # The scores for the second principal component

# Sorting the dataframe
sorted_members_by_PC2 = members_df.assign(abs_PC2_score=members_df['PC2_score'].abs()).sort_values(by='abs_PC2_score', ascending=False)

# top legislators with the largest absolute PC2 scores
top_PC2_scores = sorted_members_by_PC2.head(5)

#legislators with the smallest absolute PC2 scores
bottom_PC2_scores = sorted_members_by_PC2.tail(5)

display(top_PC2_scores[['member', 'state', 'party', 'PC2_score']])
display(bottom_PC2_scores[['member', 'state', 'party', 'PC2_score']])
```
