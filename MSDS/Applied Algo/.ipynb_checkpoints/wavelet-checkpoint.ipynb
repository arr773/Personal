```python
from typing import List
import sys


root=None

class Node:
    
    def __init__(self,bitmap,value):
        self.left=None
        self.right =None
        self.bitmap=bitmap
        self.value=value
        
class Wavelet_Tree :
    def __init__ ( self , A ) :
        self.A=A
        self.n = self.build_wavelet_tree(self.A)
        #print_wavelet_tree(self.n)
        global root
        root= self.n
        
    def build_wavelet_tree(self,A):
        bitmaparry=[]
        leftNodeArray=[]
        rightNodeArray=[]

        if len(A) ==0:
            return None

        high = - sys.maxsize -1
        low = sys.maxsize
        for i in A:
            if low >=i:
                low=i
            if high <i:
                high=i
        
       # low,high = self.find_split(A)
        if low == high:
            n= Node("X",A)
            return n

        mid= (low+high)/2

        for i in A:
            if i <= mid:
                bitmaparry.append(0)
                leftNodeArray.append(i)
            else:
                bitmaparry.append(1)
                rightNodeArray.append(i)

        if bitmaparry!="X":
            n= Node(bitmaparry,A)
            if leftNodeArray:
                n.left=self.build_wavelet_tree(leftNodeArray)
            if rightNodeArray:
                n.right=self.build_wavelet_tree(rightNodeArray)

        return n
        
    def find_split(self,A):
        high = - sys.maxsize -1
        low = sys.maxsize
        for i in A:
            if low >=i:
                low=i
            if high <i:
                high=i

        return low, high

    def print(self):
        n = self.n
        if n is None:
            return
        nodeQueue = []

        nodeQueue.append(n)
        i=0
        count=0
        listr=[]
        while(len(nodeQueue) > 0):
            count+=1

            if 2**i==count:
                
                if count !=1:
                    print()
                
                print("Level "+str(i)+" : " ,end="")
                i+=1

            print(''.join(str (b) for b in nodeQueue[0].bitmap), end=" " )
            listr.append(nodeQueue[0].bitmap)
            node = nodeQueue.pop(0)
            if node.left is not None:
                nodeQueue.append(node.left)

            if node.right is not None:
                nodeQueue.append(node.right)
        return 
            
    def calculate_RQQ(self,A,n,k,left,right,i):
        assert (0 < left and left <= right and right <= len(A))
        assert (1 <= k and k <= right - left + 1)
        print("Level "+ str(i)+": ("+str(k)+","+str(left)+","+str(right)+")")
        if (left==right or n.bitmap=="X"):
            #print("ANSWER: "+str(n.value[0]))
            return 

        left_zeros=n.bitmap[0:left-1].count(0)
        left_ones=n.bitmap[0:left-1].count(1)
        cal_bitmap= n.bitmap[left-1:right].count(0)
        direct0=n.bitmap[0:right].count(0)
        direct1=n.bitmap[0:right].count(1)

        if cal_bitmap < k:
            n=n.right
            arr=n.value
            left=left_ones+1
            right = direct1
            k=k-cal_bitmap
            i+=1
            self.calculate_RQQ(arr,n,k,left,right,i)
        else:
            n=n.left
            arr=n.value
            left = left_zeros + 1
            right = direct0
            i+=1
            self.calculate_RQQ(arr,n,k,left,right,i)
        

    def RQQ(self, k: int , left:int , right:int):
       
        assert (0 < left and left <= right and right <= len(self.A))
        assert (1 <= k and k <= right - left + 1)
        global root
        n = root
        self.calculate_RQQ(self.A,n,k,left,right,0)
        return 

wv_tree = Wavelet_Tree([6, 2, 0, 7, 9, 3, 1, 8, 5, 4])
wv_tree.print()
```
