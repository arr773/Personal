```python
from typing import List
import sys
top=None
class Node:
    def __init__(self,bitmap,value):
        self.left=None
        self.right =None
        self.bitmap=bitmap
        self.value=value
class Wavelet_Tree :
    def __init__ ( self , A ) :
        self.A=A
        self.n = self.build_wavelet_tree(self.A)
        global top
        top= self.n 
    def build_wavelet_tree(self,A):
        bitmaparry=[]
        leftNodeArray=[]
        rightNodeArray=[]
        if len(A) ==0:
            return None
        maxim = - sys.maxsize -1
        minim = sys.maxsize
        for i in A:
            if minim >=i:
                minim=i
            if maxim <i:
                maxim=i
        if minim == maxim:
            n= Node("X",A)
            return n
        mid= (minim+maxim)/2
        for i in A:
            if i <= mid:
                bitmaparry.append(0)
                leftNodeArray.append(i)
            else:
                bitmaparry.append(1)
                rightNodeArray.append(i)
        if bitmaparry!="X":
            n= Node(bitmaparry,A)
            if leftNodeArray:
                n.left=self.build_wavelet_tree(leftNodeArray)
            if rightNodeArray:
                n.right=self.build_wavelet_tree(rightNodeArray)
        return n
    def calculate_RQQ(self,A,n,k,left,right,i):
        assert (0 < left and left <= right and right <= len(A))
        assert (1 <= k and k <= right - left + 1)
        print("Level "+ str(i)+": ("+str(k)+","+str(left)+","+str(right)+")")
        if (left==right or n.bitmap=="X"):
            return 
        left_zeros=n.bitmap[0:left-1].count(0)
        left_ones=n.bitmap[0:left-1].count(1)
        cal_bitmap= n.bitmap[left-1:right].count(0)
        direct0=n.bitmap[0:right].count(0)
        direct1=n.bitmap[0:right].count(1)
        if cal_bitmap < k:
            n=n.right
            arr=n.value
            left=left_ones+1
            right = direct1
            k=k-cal_bitmap
            i+=1
            self.calculate_RQQ(arr,n,k,left,right,i)
        else:
            n=n.left
            arr=n.value
            left = left_zeros + 1
            right = direct0
            i+=1
            self.calculate_RQQ(arr,n,k,left,right,i)
    def print(self):
        n = self.n
        if n is None:
            return
        nodeQueue = []
        nodeQueue.append(n)
        i=0
        count=0
        listr=[]
        while(len(nodeQueue) > 0):
            count+=1
            if 2**i==count:
                if count !=1:
                    print()
                print("Level "+str(i)+" : " ,end="")
                i+=1
            print(''.join(str (b) for b in nodeQueue[0].bitmap), end=" " )
            listr.append(nodeQueue[0].bitmap)
            node = nodeQueue.pop(0)
            if node.left is not None:
                nodeQueue.append(node.left)
            if node.right is not None:
                nodeQueue.append(node.right)
        return 
    def RQQ(self, k: int , left:int , right:int):
        assert (0 < left and left <= right and right <= len(self.A))
        assert (1 <= k and k <= right - left + 1)
        global top
        n = top
        self.calculate_RQQ(self.A,n,k,left,right,0)
        return 

wv_tree = Wavelet_Tree([6, 2, 0, 7, 9, 3, 1, 8, 5, 4])
wv_tree.print()
wv_tree.RQQ(5,3,9)
```

```python
def find_largest_integer(num: int):
        op = list(str(num))
        q,pos=-1,-1
        s_l = len(op)-1
        for i in reversed(range(s_l-1)):
            if op[i] > op[s_l]:
                pos=i
                break       
        for i in range(s_l,pos,-1):
            if(q ==-1 and op[i]<op[pos]):
                q=i
            elif(pos >-1 and op[i]>= op[q] and op[i]<op[pos]):
                q=i
        if pos==-1:
            return int(''.join(op))
        else:
            (op[pos],op[q])=( op[q],op[pos])
        return int(''.join(op))                
            
num=182545
find_largest_integer(num)
```
