```python
class Node:
    def __init__(self,data,left=None,right=None):
        self.data=data
        self.left=left
        self.right=right
        
        
class BST:
    def __init__(self,node):
        self.root=node
        self.inor=[]
        self.pre=[]
        self.post=[]
        self.level=[]
        
        
    def __contains__(self, data,node=None):
        if node==None:
            node=self.root
        if data == node.data:
            return True
        elif data < node.data:
            if node.left:
                return self.__contains__(data,node.left)
            else:
                return False
        else:
            if node.right:
                return self.__contains__(data,node.right)
            else:
                return False
            
    
    def insert(self,data):
        if self.root==None:
            self.root=Node(data)
        else:
            new_node=Node(data)
            q=None
            p=self.root
            while(p != None):
                if data<p.data:
                    q=p
                    p=p.left
                elif data>p.data:
                    q=p
                    p=p.right
                else:
                    print("Element already in the tree")
                    return
            if data<q.data:
                    q.left=new_node
            else:
                    q.right=new_node

    
    def delete(self,data,node=None):
        if node==None:
            node=self.root
        if node == None:
            return node
        if node.data > data:
            node.left = self.delete(data,node.left)
        elif node.data < data:
            node.right = self.delete(data,node.right)  
        else:
            if node.left == None:
                p = node.right
                node = None
                return p
            elif node.right == None:
                p = node.left
                node = None
                return p 
            p = node.right
            while(p.left != None):
                p = p.left
            q = p
            node.data = q.data
            node.right = self.delete(q.data,node.right)
        return node

        
    def preorder(self,node=None):
        if node==None:
            self.pre=[]
            node=self.root
        if node:
            self.pre.append (node.data)
            if node.left:
                self.preorder(node.left)
            if node.right:
                self.preorder(node.right)
        if node==self.root:
            return self.pre
        
        
    def postorder(self,node=None):
        if node==None:
            self.post=[]
            node=self.root
        if node:
            if node.left:
                self.postorder(node.left)
            if node.right:
                self.postorder(node.right)
            self.post.append (node.data)   
        if node==self.root:
            return self.post
        
        
    def inorder(self,node=None):
        if node==None:
            self.inor=[]
            node=self.root
        if node:
            if node.left:
                self.inorder(node.left)

            self.inor.append (node.data)
            if node.right:
                self.inorder(node.right)
        if node==self.root:
            return self.inor
        
    
    def level_order(self,node=None):
        if node==None:
            node=self.root
        height = self.node_height(node)
        for i in range(1, height+1):
            self.current_level(node, i)
        return self.level
 
 
    def current_level(self,node, level):
        if node is None:
            return
        if level == 1:
            self.level.append(node.data)
        elif level > 1:
            self.current_level(node.left, level-1)
            self.current_level(node.right, level-1)

 
    def diameter(self):
        if self.root != None:
            return int(self.dia(self.root))
        else:
            return(0)
        
            
    def dia(self,node):
        if node == None:
            return(0)
        left = self.count(node.left)
        right = self.count(node.right)
        dia_left = self.dia(node.left)
        dia_right = self.dia(node.right)
        count = left + right
        if dia_left > dia_right:
            if dia_left > count:
                dia = dia_left
            else:
                dia = count
        else:
            if dia_right > count:
                dia = dia_left
            else:
                dia = count
        return dia
    
        
    def count(self,node):
        if node == None:
            return(0)
        left = self.count(node.left)
        right = self.count(node.right)
        if right > left:
            return right + 1
        else:
            return left + 1
           
    
    def height(self):
        if self.root != None:
            return (self.node_height(self.root,-1))
        else:
            return(0)

        
    def node_height(self,node,height=0):
        if node == None:
            return height
        left = self.node_height(node.left,height+1)
        right = self.node_height(node.right,height+1)
        return max(left,right)
    
    
    def max(self):
        if self.root==None:
            return None
        else:
            p=self.root
            while(p.right!=None):
                p=p.right
            return(p.data)
        pass
    
    
```
