```python
import numpy as np
from IPython.display import clear_output
def read_all_documents(path, N):
    corpus = []
    for f in range(1,N+1):
        file_path = f'{path}/{f}'
        with open(file_path,'r') as file:
            corpus.append(file.read().split())
    return corpus

path = 'pp4data/20newsgroups'
D = 200
corpus = read_all_documents(path, D)

total=0
for i in corpus:
    total+=len(i)
print(total)
N_iters = 500
K = 20
N_words = total
alpha = 5/K
beta = 0.01
v=set()
vocab=[]
for i in corpus:
    v.update(set(i))
for i in (v):
    vocab.append(i)
V = len(vocab)
vocab
```

```python
import random
pi_n = np.random.permutation(total)

w_n=[]
for i in corpus:
    for j in i:
        w_n.append(vocab.index(j))
w_n = np.array(w_n)
        
d_n=[]
for i in range(len(corpus)):
    for j in range(len(corpus[i])):
        d_n.append(i)
d_n = np.array(d_n)

z_n=[]
for i in range(total):
    z_n.append(random.randint(0,K-1))
z_n = np.array(z_n)
    
C_d = np.zeros((D, K))
C_t = np.zeros((K,V))
#print(w_n)
for d in range(C_d.shape[0]):
    for t in range(C_d.shape[1]):
        C_d[d,t] = np.sum(z_n[np.where(d_n==d)[0]]==t)

for t in range(C_t.shape[0]):
    for w in range(C_t.shape[1]):
        C_t[t,w] = np.sum(w_n[np.where(z_n==t)[0]]==w)

size3 = 1*K
#P = np.array([0]*size3).reshape(1, K)
P = np.zeros(K)

#P.shape
#sum(P)

for i in range(N_iters):
    for n in range (N_words):
        word = w_n[pi_n[n]]
        topic = z_n[pi_n[n]]
        #print(topic)
        doc = d_n[pi_n[n]]
        #print(doc)
        C_d[doc][topic] = C_d[doc][topic] - 1
        #print(C_d[doc][topic])
        C_t[topic][word] = C_t[topic][word] - 1
        for k in range(K):
            term1 = ((C_t[k,word] + beta)/(V*beta + np.sum(C_t[k,:])))
            term2 = ((C_d[doc,k] + alpha)/(K*alpha + np.sum(C_d[doc,:])))
            P[k] = term1*term2
        pnorm = sum(P)
        pnorm = pnorm if pnorm>0 else 1
        P = [pp/pnorm for pp in P]
        topic = np.random.choice(np.arange(K), p=P)
        z_n[pi_n[n]] = topic
        C_d[doc][topic] = C_d[doc][topic] + 1
        C_t[topic][word] = C_t[topic][word] + 1
```

```python
def newton(newX,Y, data):
    N = newX.shape[0]
    d = newX.shape[1]
    train_size = int((2/3) * len(data))
    xtrain = newX[:train_size]
    ytrain = Y[:train_size]
    xtest = newX[train_size:]
    ytest = Y[train_size:]
   # W_old = np.ones(xtrain.shape[1])
    W_old = np.zeros(xtrain.shape[1])
    alpha = 0.1
    iteration = 0
    weight_vector = {}
    errors = {}
    time_diff = []
    initial_time = time.time()
    while True:
        a = np.dot(xtrain, W_old)
        y = 1/(1 + np.exp(-a))
        #print(y.shape)
        R = np.diag(y*(1-y))
        #print(R.shape,R)
        term_1 = alpha*np.identity(d) + np.dot(np.dot(xtrain.T,R),xtrain)
        term_2 = np.dot(xtrain.T, (y-ytrain)) + alpha*W_old
        W_new = W_old - np.dot(np.linalg.inv(term_1),term_2)
        iteration+=1
        #W_old = W_new
        ts = time.time()
        weight_vector[ts] = W_new 
        
        if ((np.linalg.norm(W_new-W_old)**2)/(np.linalg.norm(W_old)**2)<0.001) or iteration>100:
            break
        else:
            W_old = W_new
    for ts in weight_vector:
                
                
        # SN calculation
        a = np.dot(xtrain, weight_vector[ts])
        y = 1/(1 + np.exp(-a))
        summation = np.zeros((d,d))
        for i in range(y.size):
            summation+= (y[i]*(1-y[i]))*np.dot(xtrain[i,:], xtrain[i,:])
            
        SN = alpha*np.identity(d) + summation
        #predictions

        pred = np.zeros(xtest.shape[0])
        for i in range(pred.size):
            phi = xtest[i,:]
            mu = np.dot(weight_vector[ts].T, phi)
            sigma_sq = np.dot(np.dot(phi.T, SN), phi)
            a = 1/np.sqrt(1 + (np.pi*sigma_sq/8))
            pred[i] = 1/(1 + np.exp(-(a*mu)))
                     
        prediction = np.where(pred>=0.5,1,0)
        error = np.sum(np.logical_xor(ytest, prediction))
        test_error = error/ytest.size
        errors[ts-initial_time] = test_error
        time_diff.append(ts-initial_time)   
            
    return errors, weight_vector, time_diff
```
