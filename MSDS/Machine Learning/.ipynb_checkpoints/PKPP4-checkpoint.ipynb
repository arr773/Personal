```python
import numpy as np
from IPython.display import clear_output
def read_all_documents(path, N):
    corpus = []
    for f in range(1,N+1):
        file_path = f'{path}/{f}'
        with open(file_path,'r') as file:
            corpus.append(file.read().split())
    return corpus

path = 'pp4data/20newsgroups'
D = 200
corpus = read_all_documents(path, D)

total=0
for i in corpus:
    total+=len(i)
print(total)
N_iters = 500
K = 20
N_words = total
alpha = 5/K
beta = 0.01
v=set()
vocab=[]
for i in corpus:
    v.update(set(i))
for i in (v):
    vocab.append(i)
V = len(vocab)
vocab
```

```python
import random
pi_n = np.random.permutation(total)

w_n=[]
for i in corpus:
    for j in i:
        w_n.append(vocab.index(j))
w_n = np.array(w_n)
        
d_n=[]
for i in range(len(corpus)):
    for j in range(len(corpus[i])):
        d_n.append(i)
d_n = np.array(d_n)

z_n=[]
for i in range(total):
    z_n.append(random.randint(0,K-1))
z_n = np.array(z_n)
    
C_d = np.zeros((D,K))
C_t = np.zeros((K,V))
#print(w_n)
for d in range(C_d.shape[0]):
    for t in range(C_d.shape[1]):
        C_d[d,t] = np.sum(z_n[np.where(d_n==d)[0]]==t)

for t in range(C_t.shape[0]):
    for w in range(C_t.shape[1]):
        C_t[t,w] = np.sum(w_n[np.where(z_n==t)[0]]==w)

size3 = 1*K
#P = np.array([0]*size3).reshape(1, K)
P = np.zeros(K)

#P.shape
#sum(P)

for i in range(N_iters):
    for n in range (N_words):
        word = w_n[pi_n[n]]
        topic = z_n[pi_n[n]]
        #print(topic)
        doc = d_n[pi_n[n]]
        #print(doc)
        C_d[doc][topic] = C_d[doc][topic] - 1
        #print(C_d[doc][topic])
        C_t[topic][word] = C_t[topic][word] - 1
        for k in range(K):
            term1 = ((C_t[k,word] + beta)/(V*beta + np.sum(C_t[k,:])))
            term2 = ((C_d[doc,k] + alpha)/(K*alpha + np.sum(C_d[doc,:])))
            P[k] = term1*term2
        pnorm = sum(P)
        pnorm = pnorm if pnorm>0 else 1
        P = [pp/pnorm for pp in P]
        topic = np.random.choice(np.arange(K), p=P)
        z_n[pi_n[n]] = topic
        C_d[doc][topic] = C_d[doc][topic] + 1
        C_t[topic][word] = C_t[topic][word] + 1
```

```python
def bagOfWordsRepresentation(corpus):
    corpus_words = [w for d in corpus for w in d]
    
    # word - index mapping
    dict_words = {}
    iter = 0
    for w in corpus_words:
        if w not in dict_words:
            dict_words[w] = iter
            iter+=1
    V = len(dict_words)
    
    # calculating the indices vector D and W
    Dvec = []

    for i,doc in enumerate(corpus):
        Dvec.extend([i for w in range(len(doc))])
    
    Dvec = np.array(Dvec)
    Wvec = []

    for w in corpus_words:
        Wvec.append(dict_words[w])
    
    Wvec = np.array(Wvec)

    # calculating doc_word representation

    doc_words = np.zeros((D,V))

    for d in range(doc_words.shape[0]):
        for w in range(doc_words.shape[1]):
            doc_words[d,w] = np.sum(Wvec[np.where(Dvec==d)[0]]==w)
    
    return doc_words
```

```python
def discriminative(X, Y):
    N = X.shape[0]
    d = X.shape[1]
    N_test = int(N/3)
    N_train = N-N_test
    all_index = [i for i in range(N)]
    discriminative_result = {}
    logisticR_X = np.c_[np.ones(N),X]
    for times in range(1, 31):
        test_index = random.sample(all_index, N_test)
        train_index = [x for x in all_index if x not in test_index]
        xtrain = logisticR_X[train_index,:]
        ytrain = Y[train_index]
        xtest = logisticR_X[test_index,:]
        ytest = Y[test_index]
        for pv in partition_values:
            alpha = 0.1
            pv_N = int(pv*N_train)
            xpv = xtrain[:pv_N,]
            ypv = ytrain[:pv_N]
            W_old = np.zeros(xpv.shape[1])
            #W_new = np.matrix(np.zeros(xpv.shape[1]))
            iteration = 0
            while True:
                a = np.dot(xpv, W_old)
                y = 1/(1 + np.exp(-a))
                R = np.diag(y*(1-y))
                term_1 = alpha*np.identity(d+1) + np.dot(np.dot(xpv.T,R),xpv)
                term_2 = np.dot(xpv.T, (y-ypv)) + alpha*W_old
                W_new = W_old - np.dot(np.linalg.inv(term_1),term_2)

                #W_old=W_new
                iteration+=1
                if ((np.linalg.norm(W_new-W_old)**2)/(np.linalg.norm(W_old)**2)<0.001) or iteration>100:
                    break
                else:
                    W_old = W_new

            # SN calculation
            a = np.dot(xpv, W_new)
            y = 1/(1 + np.exp(-a))
            summation = np.zeros((d+1,d+1))
            for i in range(y.size):
                summation+= (y[i]*(1-y[i]))*np.dot(xpv[i,:], xpv[i,:])
            
            SN = alpha*np.identity(d+1) + summation
            
            #predictions

            pred = np.zeros(xtest.shape[0])
            for i in range(pred.size):
                phi = xtest[i,:]
                mu = np.dot(W_new.T, phi)
                sigma_sq = np.dot(np.dot(phi.T, SN), phi)
                a = 1/np.sqrt(1 + (np.pi*sigma_sq/8))

                pred[i] = 1/(1 + np.exp(-(a*mu)))
            
            prediction = np.where(pred>=0.5,1,0)
            error = np.sum(np.logical_xor(ytest, prediction))
            test_error = error/ytest.size
            
            
            discriminative_result[(str(times),str(pv))] = test_error
    return discriminative_result
```

```python
def average_error(dr):
    discriminative_avg = {}

    for pv in partition_values:
        arr = np.array([dr[(str(e),str(pv))] for e in range(1,31)])
        discriminative_avg[str(pv)] = (np.mean(arr), np.std(arr))
    return discriminative_avg
```

```python
start1 = time.time()
prediction = discriminative(X1,Y)
end1 = time.time()
diff1 = end1-start1
```
