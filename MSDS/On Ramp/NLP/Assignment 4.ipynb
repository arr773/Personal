```python
import json
data=[]
with open('./streamingData.json', 'r') as jsonFile:
    for line in jsonFile:
        data.append(json.loads(line))
```

```python
import re
import string
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
tweet = []
for item in data:
    if "text" in item.keys():
        tweet.append(item["text"])
```

```python
#removing hashtags
for i in range(len(tweet)):
    tweet[i] = re.sub(r'#\w+', '',tweet[i] )
    tweet[i]=tweet[i].lower()
```

```python
#remove apostrophe's
for i in range(len(tweet)):
    tweet[i] = re.sub(r"it's", "it is", tweet[i])
    tweet[i] = re.sub(r"won't", "will not", tweet[i])
    tweet[i] = re.sub(r"can't", "cannot", tweet[i])
    tweet[i] = re.sub(r"don't", "do not", tweet[i])
    tweet[i] = re.sub(r"i'm", "i am", tweet[i])
    tweet[i] = re.sub(r"he's", "he is", tweet[i])
    tweet[i] = re.sub(r"she's", "she is", tweet[i])
    tweet[i] = re.sub(r"they're", "they are", tweet[i])
    tweet[i] = re.sub(r"we're", "we are", tweet[i])
    tweet[i] = re.sub(r"you're", "you are", tweet[i])
    tweet[i] = re.sub(r"n\'t", " not", tweet[i])
    tweet[i] = re.sub(r"\'re", " are", tweet[i])
    tweet[i] = re.sub(r"\'s", " is", tweet[i])
    tweet[i] = re.sub(r"\'d", " would", tweet[i])
    tweet[i] = re.sub(r"\'ll", " will", tweet[i])
    tweet[i] = re.sub(r"\'t", " not", tweet[i])
    tweet[i] = re.sub(r"\'ve", " have", tweet[i])
    tweet[i] = re.sub(r"\'m", " am", tweet[i])
```

```python
#remove punctuations
for i in range(len(tweet)):
    tweet[i] = re.sub(r'[^\w\s?!]', '', tweet[i])
```

```python
#removing patterns
for i in range(len(tweet)):
    tweet[i] = re.sub(r" gud ", " good ", tweet[i])
    tweet[i] = re.sub(r" awsm ", " awesome ", tweet[i])
    tweet[i] = re.sub(r" luv ", " love ", tweet[i])
    tweet[i] = re.sub(r" ur ", " your ", tweet[i])
    tweet[i] = re.sub(r" im ", " i am ", tweet[i])
    tweet[i] = re.sub(r" dont ", " do not ", tweet[i])
    tweet[i] = re.sub(r" wont ", " will not ", tweet[i])
    tweet[i] = re.sub(r" cant ", " can not ", tweet[i])
    tweet[i] = re.sub(r" shouldnt ", " should not ", tweet[i])
    tweet[i] = re.sub(r" wouldnt ", " would not ", tweet[i])
    tweet[i] = re.sub(r" couldve ", " could have ", tweet[i])
    tweet[i] = re.sub(r" mustve ", " must have ", tweet[i])
    tweet[i] = re.sub(r" aint ", " am not ", tweet[i])
    tweet[i] = re.sub(r" isnt ", " is not ", tweet[i])
    tweet[i] = re.sub(r" werent ", " were not ", tweet[i])
    tweet[i] = re.sub(r" hasnt ", " has not ", tweet[i])
    tweet[i] = re.sub(r" havent ", " have not ", tweet[i])
    tweet[i] = re.sub(r" hadnt ", " had not ", tweet[i])
    tweet[i] = re.sub(r" wanna ", " want to ", tweet[i])
```

```python
#sentiment analysis
import nltk
nltk.download('omw-1.4')
nltk.download('sentiwordnet')
from nltk.corpus import sentiwordnet as swn

def get_sentiment(word):
    synsets = list(swn.senti_synsets(word))
    if synsets:        
        synset = synsets[0]
        return synset.pos_score() - synset.neg_score()
    else:
        return 0
def calculate_tweet_polarity(tweet):
    words=tweet.split()
    sentiment=0
    for w in words:
        sentiment+=get_sentiment(w)
    if len(words)<=0:
        sentiment=0
    return sentiment

for i in tweet:
    polarity=calculate_tweet_polarity(i)
    print(i,polarity,"\n")
```
