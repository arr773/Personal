```python
from selenium import webdriver
import chromedriver_autoinstaller
from selenium.webdriver.common.by import By
import time
chromedriver_autoinstaller.install()  
driver = webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/ml/datasets.php")
time.sleep(5)
data1=[]
rows=len(driver.find_element(By.XPATH, "/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr").text)
for i in range(1,rows+1):
    data1.append(driver.find_element(By.XPATH, "/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]").text)
# print(data1)
driver.close()
```

```python
data2=[]
for i in data1:
    data2.append(i.split("\n"))
fields=data2[0]
data2.sort(key=lambda x:x[len(x)-1],reverse=True)
rows=data2[1:]
```

```python
import csv
with open('assignment5', 'w') as f:
      
    # using csv.writer method from CSV package
    write = csv.writer(f)
      
    write.writerow(fields)
    write.writerows(rows)
f.close()
```
